# Creating merged DataFrames that would come in handy later:

Used left-join on all merges so that we don't lose important information from the datasets on the left where the datasets on the right don't have corresponding information.

```python
customers_order_df = pd.merge(customers_df, orders_df, on='customer_id', how='left')
customers_order_df
```

```python
for col in ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']:
    customers_order_df[col] = pd.to_datetime(customers_order_df[col])
```

```python
customers_order_items_df = pd.merge(customers_order_df, order_items_df, on='order_id', how='left')
customers_order_items_df
```

# Exploring the dataset

Our dataset contains orders made between **4th September, 2016** and **17th October, 2018.**

A total of about **773 days.**

We have a total of **99441 unique orders** in our dataset.

Some *orders* are *missing* from the **order_items** dataset.

These orders number **775** in total and further investigation shows that their order statuses are mostly either *'unavailable'* or *'canceled'*.

*These orders being missing from the order_items dataset means that we can't find the products, product prices, freight values, sellers nor product categories linked to these orders.*

```python
missing_orders = []
available_orders = []
set_1 = set(orders_df['order_id'].tolist())
set_2 = set(order_items_df['order_id'].tolist())


for order in set_1:
    if order in set_2:
        available_orders.append(order)
    else:
        missing_orders.append(order)
```

```python
len(available_orders)    #this matches the count of unique order_ids in the order_items dataset
```

```python
len(missing_orders)    #this is the number of order_ids in the orders dataset that are missing from the order_items dataset
```

```python
missing_orders_df = pd.DataFrame(columns=orders_df.columns)
for order in missing_orders:
    temp_df = orders_df[orders_df['order_id']==order]
    missing_orders_df = missing_orders_df.append(temp_df, ignore_index=True)
missing_orders_df['order_status'].value_counts()
```

**Time Series Analysis of the orders:**

```python
time_series_df = customers_order_items_df.set_index('order_purchase_timestamp')
```

```python
#value of orders with respect to time
```

```python
price_time_series_df = time_series_df[['price', 'freight_value']]
price_time_series_df = price_time_series_df.resample('D').sum()
price_time_series_df
```

```python
price_time_series_df['price'].sum()    #should still be equal to the total value before resampling
```

```python
price_moving_average = price_time_series_df['price'].rolling(7).mean()
```

```python
plt.close()
```

```python
plt.figure(figsize=(16, 9))
plt.plot(price_time_series_df.index, price_time_series_df['price'])
plt.plot(price_moving_average, color='r');
```

We see a spike in money earned on a particular day towards the end of the year.

Looking at this a bit further;

```python
price_time_series_df['price'].idxmax()
```

The date of this spike in revenue is 24th November, 2017. Incidentally, this is also the date for Black Friday that year. The Black Friday sales on the Olist platform brought in a huge surge in orders which in turn corresponds to a huge surge in revenue.

link to verify date: https://www.google.com/search?client=firefox-b-d&q=black+friday+2017

There is a steady general increase in revenue generated with time. We have a dip and flatline at the end of our graph because our has no price information for most of the orders carried out after August 2018.

Revenue dips around the turn of the year 2017 - 2018 with two other noticeable dips in June and July 2018.

```python
#checking how long it usually takes for orders to get approved after they've been made.

((customers_order_df['order_approved_at'] - customers_order_df['order_purchase_timestamp']).dt.days).value_counts().sort_index()
```

As we can see, most orders are approved less than 24hrs after they've been made.
We see a few orders that have taken up to a month to get approved and a couple that took even longer.

```python
#checking how long it usually takes for orders to get delivered to the customer.

((customers_order_df['order_delivered_customer_date'] - customers_order_df['order_purchase_timestamp']).dt.days).value_counts().sort_index()
```

```python
(((customers_order_df['order_delivered_customer_date'] - customers_order_df['order_purchase_timestamp']).dt.days).value_counts().sort_index()).plot(figsize=(32,18), kind='bar');
```

As we can see, most orders are delivered just a few days after they are placed.
A few orders take months to complete though.

```python
#checking the difference between the estimated delivery customers are given and when they actually receive their orders.

(((customers_order_df['order_estimated_delivery_date'] - customers_order_df['order_delivered_customer_date']).dt.days).value_counts().sort_index()).plot(figsize=(32,18), kind='bar');
```

```python
((customers_order_df['order_estimated_delivery_date'] - customers_order_df['order_delivered_customer_date']).dt.days).value_counts().head(50)
```

Most orders are completed well before estimated delivery dates. The frequency of late deliveries might be a cause for concern though as it could affect customer satisfaction.
